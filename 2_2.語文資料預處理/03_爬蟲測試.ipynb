# Beautiful Soup 測試

```python
# 載入相關套件
import requests
from bs4 import BeautifulSoup
```

```python
# 送出 request，取得 response
url = 'https://sports.ltn.com.tw/nba'
response = requests.get(url)
contents = response.content
contents
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(contents, 'html.parser')

print(soup.prettify())
```

```python
soup.title #----->網頁標籤
```

```python
soup.title.name
```

```python
soup.title.string
```

```python
for link in soup.find_all('li', {'data-page':'1'}):
    a = link.find('h3')
    print(a.text.strip()) #a['href'])
```
