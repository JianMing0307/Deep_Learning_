```python
import jieba
import collections
```

```python
def read_text(file_path):
    # 1.開啟檔案
    with open(file_path, "r", encoding='utf-8') as f: #----->先看mode,再從指定的路徑取出排列到暫存區(f)
        data = f.readlines() #----->獨到\n就停下來
        word_list = []
        # 2. 逐行讀取 
        for line in data:
            line = line.strip() #----->去前後空白
            words = jieba.cut(line)
            for word in words:
                word_list.append(word) #----->2維
    # 3. 統計word_list中每個單詞出現的次數
    counter = collections.Counter(word_list).most_common()
    print(counter)
```

```python
file_path = "./news2.txt"
read_text(file_path)
```
